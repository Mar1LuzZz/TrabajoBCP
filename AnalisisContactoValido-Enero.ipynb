```python
DF_final=pd.DataFrame(columns=SCC_Febrero.columns.tolist())
```

```python
DF_final
```

```python
for i in telefonos: 
    df_temp=SCC_Febrero[SCC_Febrero['TELEFONO_GESTION']==i]
    df_temp_CV=df_temp[df_temp['DESCRIPCION_GRUPO_TIPIFICACION']=='CONTACTO V√ÅLIDO']
    if df_temp_CV.empty==True:
        df_concatenar=df_temp
    else: 
        df_concatenar=df_temp_CV
    DF_final=pd.concat([DF_final, df_concatenar], ignore_index=True)
```
